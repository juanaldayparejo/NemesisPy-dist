# Defining the Stellar Spectrum

In this notebook we provide some examples showing how NemesisPy can be used to define the stellar spectrum for an atmospheric retrieval. The main input file required for NemesisPy to define the solar spectrum is the $.sol$ file, which just includes one line with the name of the file where the solar spectrum is stored. Note this file must be defined in a specific format and it is expected to be stored in $Data/stellar$.

```python
from NemesisPy import *
```

## Reading the stellar spectrum from the input file

NemesisPy includes one function to read the $.sol$ file and automatically load the stellar spectrum. In this example, we will read the solar spectrum in both wavelength and wavenumber units, using the $.sol$ files in the example directory.

```python
Stellar = Stellar_0()
Stellar.read_sol('solspec_wl')

fig,ax1 = plt.subplots(1,1,figsize=(10,3))
ax1.plot(Stellar.VCONV,Stellar.SOLSPEC)
ax1.grid()
if Stellar.ISPACE==0:
    ax1.set_xlabel('Wavenumber (cm$^{-1}$)')
    ax1.set_ylabel('Spectral luminosity (W (cm$^{-1}$)$^{-1}$)')
elif Stellar.ISPACE==1:
    ax1.set_xlabel('Wavelength ($\mu$m)')
    ax1.set_ylabel('Spectral luminosity (W $\mu$m$^{-1}$)')
plt.tight_layout()
```

```python
Stellar = Stellar_0()
Stellar.read_sol('solspec_wn')

fig,ax1 = plt.subplots(1,1,figsize=(10,3))
ax1.plot(Stellar.VCONV,Stellar.SOLSPEC)
ax1.grid()
if Stellar.ISPACE==0:
    ax1.set_xlabel('Wavenumber (cm$^{-1}$)')
    ax1.set_ylabel('Spectral luminosity (W (cm$^{-1}$)$^{-1}$)')
elif Stellar.ISPACE==1:
    ax1.set_xlabel('Wavelength ($\mu$m)')
    ax1.set_ylabel('Spectral luminosity (W $\mu$m$^{-1}$)')
plt.tight_layout()
```

## Calculating the solar irradiance at a given distance

Once the stellar spectrum has been read, the stellar flux or irradiance at a given distance to the star can be easily calculated following

\begin{equation}
F(\lambda,d) = \dfrac{P(\lambda)}{4 \pi d^2},
\end{equation}

where $P(\lambda)$ is the stellar spectral luminosity and $d$ is the distance to the star.

```python
Stellar = Stellar_0()
Stellar.read_sol('solspec_wl')

#Defining the distance to star
Stellar.DIST = 1.5 #Astronomical Units (Sun-Mars distance)

#Calculating the irradiance
Stellar.calc_solar_flux()

#Plotting the solar irradiance
fig,ax1 = plt.subplots(1,1,figsize=(10,3))
ax1.plot(Stellar.VCONV,Stellar.SOLFLUX)
ax1.grid()
if Stellar.ISPACE==0:
    ax1.set_xlabel('Wavenumber (cm$^{-1}$)')
    ax1.set_ylabel('Spectral irradiance (W cm$^{-2}$ (cm$^{-1}$)$^{-1}$)')
elif Stellar.ISPACE==1:
    ax1.set_xlabel('Wavelength ($\mu$m)')
    ax1.set_ylabel('Spectral irradiance (W cm$^{-2}$ $\mu$m$^{-1}$)')
plt.tight_layout()
```
